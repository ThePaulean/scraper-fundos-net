### Encontrar as tabelas

Neste notebook iremos percorrer os documentos PDF e tentar encontrar as tabelas de CRI dentro deles

```python
import pandas as pd

df = pd.read_excel("files/fnet_docs_filtered.xlsx", index_col=0)
df = df.dropna(axis=1, how="all")
df
```

```python
import re


regex_pattern = re.compile(r"\b(CRI|CRA)\b")
```

```python
import fitz
import numpy as np

for idx, row in df.iterrows():
    # print(idx)
    file = row["doc_path"]
    doc = fitz.open(file)
    pages = []
    for p in doc.pages():
        p : fitz.Page
        text = p.get_text()
        if regex_pattern.search(text):
            # print("Found")
            pages.append(p.number)

    if pages:
        print(idx, pages)
        df.at[idx, "has_cri_cra"] = True
        df.at[idx, "pages"] = ",".join(list(map(str, pages)))
    else:
        df.at[idx, "has_cri_cra"] = False
        df.at[idx, "pages"] = np.nan # ",".join(list(map(str, pages)))
        

df
```
