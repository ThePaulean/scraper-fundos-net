```python
import polars as pl

df_docs = pl.read_excel("files/fnet_docs.xlsx")
df_docs
```

```python
df_docs_pd = df_docs.to_pandas()
df_docs_pd
```

```python
# Treating the dataframe in pandas

df_docs_pd = df_docs_pd.iloc[:, 1:]
df_docs_pd = df_docs_pd.dropna(axis=1, how="all")
df_docs_pd = df_docs_pd[df_docs_pd["status"] == "AC"]
df_docs_pd = df_docs_pd[["id", "descricaoFundo", "dataReferencia", "dataEntrega", "nomePregao", "url", "doc_path"]]
df_docs_pd
```

Vamos pegar uma amostra de 200 documentos

```python
df_analysis = df_docs_pd.copy()
df_analysis = pl.from_pandas(df_analysis)
df_analysis = df_analysis[:200]
df_analysis
```

```python
import tabula
import fitz
from tqdm import tqdm

updated_rows = []
for row in tqdm(df_analysis.iter_rows()):
    updated_row = list(row)
    doc_path = row[-1]
    doc = fitz.Document(doc_path)
    pages_w_table = []
    for p_num in range(doc.page_count):
        try:
            dfs = tabula.read_pdf(doc_path, pages=p_num+1)
            if dfs:
                pages_w_table.append(p_num+1)
        except Exception as e:
            print(e)
    updated_row.append(",".join(map(str, pages_w_table)))
    updated_rows.append(updated_row)

df_table_pages = pl.DataFrame(updated_rows, df_analysis.columns + ["pages_w_table"])
df_table_pages
```

```python
df_table_pages.write_excel("Tables_finding.xlsx")
```
