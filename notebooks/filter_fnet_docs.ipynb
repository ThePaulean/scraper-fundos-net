### Objetivo do Notebook

Neste notebook irei formatar o arquivo XLSX baseado nas informações que preenchi após verificar 200 documentos.

Vamos filtrar os fundos que possuem tabelas de CRI e verificar, dos fundos que não validei, quais podem, potencialmente, ter alguma tabela de CRI, para caso precisemos de mais dados.

```python
# Abrindo o documento
import pandas as pd

df = pd.read_excel("files/fnet_docs.xlsx", index_col=0)
print([col for col in df.columns])
df
```

---
### Limpeza dos casos que não possuem CRI

Aqui, iremos criar dois DataFrames a partir do inicial. Um deles irá conter os dados que possuem CRI, o outro não.

Esses DataFrames serão auxiliares para uma filtração mais profunda em seguida.

---

```python
df_aux_has_cri = df[df["possui_cri"] == "Sim"]
df_aux_lacks_cri = df[df["possui_cri"] == "Não"]

###

# Checamos para ver se tem alguma interseção
df_intersec = pd.concat([
    df_aux_has_cri[df_aux_has_cri["descricaoFundo"].isin(df_aux_lacks_cri["descricaoFundo"])],
    df_aux_lacks_cri[df_aux_lacks_cri["descricaoFundo"].isin(df_aux_has_cri["descricaoFundo"])],
]).sort_values(by="descricaoFundo")
df_intersec
```

```python
import os
import time

for idx, row in df_intersec.iterrows():
    print(row["descricaoFundo"])
    # os.startfile(row["doc_path"])
    # time.sleep(3)
```

---

### Capturar casos positivos

Agora que limpamos as interseções que estavam ou não erradas, vamos pegar do DataFrame original somente aqueles casos cujo fundo está presente no DataFrame positivo.

---

```python
df_filter = df[df["descricaoFundo"].isin(df_aux_has_cri["descricaoFundo"])]
df_filter = df_filter.reset_index(drop=True)
df_filter
```
